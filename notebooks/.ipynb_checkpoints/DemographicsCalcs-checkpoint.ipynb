## Calculations dataframe for Demographics Columns by Percent Passing. Grades 3, 4, and 5.

### Import dependencies and modules

```python
import pandas as pd
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
```

### Open demographics data, Grade 3

```python
demographics3 = pd.read_csv('../cleandata/DemographicsHABWGrade3.csv', header=0)
demographics3.head(3)
```

### For data processing across rows, create a new dataframe with only the number data and district type

```python
demographics3.columns
```

```python
newdg3_df = demographics3[['DNAME', 'CNAME', 'District Type',
       'Below Grade Level Hispanic', 'Below Grade Level Asian',
       'Below Grade Level Black', 'Below Grade Level White',
       'Approach Grade Level Hispanic', 'Approach Grade Level Asian',
       'Approach Grade Level Black', 'Approach Grade Level White',
       'Meet Grade Level Hispanic', 'Meet Grade Level Asian',
       'Meet Grade Level Black', 'Meet Grade Level White',
       'Master Grade Level Hispanic', 'Master Grade Level Asian',
       'Master Grade Level Black', 'Master Grade Level White']]
```

```python
newdg3_df.head()
```

### Make a new dataframe for the pass information, so as to not mess up newdg3_df in place.

```python
newdg3_pass_df = newdg3_df
```

### How to calculate a reliable measure of student performance??? Try percent of each ethnic category that has Meet or Master. For each ethnic category, sum students in Meet and Master, then divide by total number of students in ethnic category. Note that some schools have no students in a category, so use try, except error handling. When I tried to bundle all of the newdg3_pass_df.loc arguments into one try, the calculations did not work properly. Individual try, except did work!

```python
for index, row in newdg3_pass_df.iterrows():
    
    try:
        newdg3_pass_df.loc[index, '%Pass Hispanic'] = (row[11]+row[15])/(row[3]+row[7]+row[11]+row[15])
    except: ZeroDivisionError
        
    try:
        newdg3_df.loc[index, '%Pass Asian'] = (row[12]+row[16])/(row[4]+row[8]+row[12]+row[16])
    except: ZeroDivisionError
        
    try:   
        newdg3_df.loc[index, '%Pass Black'] = (row[13]+row[17])/(row[5]+row[9]+row[13]+row[17]) 
    except: ZeroDivisionError
        
    try:  
        newdg3_pass_df.loc[index, '%Pass White'] = (row[14]+row[18])/(row[6]+row[10]+row[14]+row[18])   
    except: ZeroDivisionError
        
```

### Check if calculations worked

```python
newdg3_pass_df.head(10)
```

### Another check for calculations. Appear to have worked.

```python
newdg3_pass_df.count()
```

### Make a new dataframe for final analysis and plotting of %Pass data based on Independent or Charter school type.

```python
pass3_df = newdg3_pass_df[['District Type', '%Pass White', '%Pass Black', '%Pass Hispanic', '%Pass Asian']]
pass3_df.head(10)
```

### Strategy: If I remove NaN rows, most data will be lost (Asian column has only 886 values). Break up into four different dataframes for removing NaN rows and plotting. NOTE: Check on NaN values indicated no data in the original data for these demographic populations. ZeroDivisionError gives NaN and not informative for analysis.

```python
pass3_white = pass3_df[['District Type', '%Pass White']]
pass3_black = pass3_df[['District Type', '%Pass Black']]
pass3_hispanic = pass3_df[['District Type', '%Pass Hispanic']]
pass3_asian = pass3_df[['District Type', '%Pass Asian']]
```

```python
pass3_asian.count()
```

### dropna for dataframes

```python
pass3_white_dpna = pass3_white.dropna(how='any')
pass3_black_dpna = pass3_black.dropna(how='any')
pass3_hispanic_dpna = pass3_hispanic.dropna(how='any')
pass3_asian_dpna = pass3_asian.dropna(how='any')
```

```python
pass3_hispanic_dpna.count()
```

### Create Independent and Charter School dataframes for calculations.

```python
pass3white_cht = pass3_white_dpna.loc[pass3_white_dpna['District Type'] == 'CHARTER']
pass3white_inp = pass3_white_dpna.loc[pass3_white_dpna['District Type'] == 'INDEPENDENT']

pass3black_cht = pass3_black_dpna.loc[pass3_black_dpna['District Type'] == 'CHARTER']
pass3black_inp = pass3_black_dpna.loc[pass3_black_dpna['District Type'] == 'INDEPENDENT']

pass3hisp_cht = pass3_hispanic_dpna.loc[pass3_hispanic_dpna['District Type'] == 'CHARTER']
pass3hisp_inp = pass3_hispanic_dpna.loc[pass3_hispanic_dpna['District Type'] == 'INDEPENDENT']

pass3asian_cht = pass3_asian_dpna.loc[pass3_asian_dpna['District Type'] == 'CHARTER']
pass3asian_inp = pass3_asian_dpna.loc[pass3_asian_dpna['District Type'] == 'INDEPENDENT']
```

### TTest for data. Convert all of these data to vectors, then apply TTest through Scipy.

```python
pass3cht_vector = []
for index, row in pass3white_cht.iterrows():
    pass3cht_vector.append(row[1])

pass3inp_vector = []
for index, row in pass3white_inp.iterrows():
    pass3inp_vector.append(row[1])

sp.ttest_ind(pass3cht_vector, pass3inp_vector, equal_var=True)
```

### Calculate means

```python
ht = [pass3white_inp.mean()[0], pass3white_cht.mean()[0], pass3black_inp.mean()[0], pass3black_cht.mean()[0], pass3hisp_inp.mean()[0], pass3hisp_cht.mean()[0], pass3asian_inp.mean()[0], pass3asian_cht.mean()[0]]
ht
```

### Calculate errors

```python
errors = [pass3white_inp.sem()[0], pass3white_cht.sem()[0], pass3black_inp.sem()[0], pass3black_cht.sem()[0], pass3hisp_inp.sem()[0], pass3hisp_cht.sem()[0], pass3asian_inp.sem()[0], pass3asian_cht.sem()[0]]
errors
```

### Create bar graph using data from ht(means) and errors(sem) data.

```python
plt.figure(figsize=(9,6))
x_axis = [i for i in range(len(ht))]

barfig = plt.bar(x_axis, height=ht, yerr=errors, color='cyan', edgecolor='k', align='center', width=0.75)
barfig[0].set_color('yellow')
barfig[0].set_edgecolor('k')
barfig[2].set_color('yellow')
barfig[2].set_edgecolor('k')
barfig[4].set_color('yellow')
barfig[4].set_edgecolor('k')
barfig[6].set_color('yellow')
barfig[6].set_edgecolor('k')

plt.ylim(0,0.6)

plt.title('Demographics Grade 3: % Students Passing 2019 STAAR Test', fontsize=16, fontweight='bold')
plt.xlabel('School Type', fontsize=12, fontweight='bold')
plt.ylabel('%Pass (Meet or Master Grade Level)', fontsize=12, fontweight='bold')

plt.xticks(x_axis, ['Independent(White)', 'Charter(White)', 'Independent(Black)', 'Charter (Black)', 'Independent(Hispanic)', 'Charter (Hispanic)', 'Independent (Asian)', 'Charter (Asian)'], ha='right', fontsize = 12, rotation=45)

plt.tight_layout()

#Save figure
plt.savefig('../reports/BarDemographicsGrade3', dpi=300)
```
